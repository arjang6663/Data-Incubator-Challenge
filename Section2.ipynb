### If you plot the number of hotspot activations for each month, you'll notice a general increase but then a precipitous drop after June 2018. Using a linear estimate for the number of monthly activations, what is rate of increase in monthly activations? Only consider data before July 1, 2018 and set the start date as the earliest date of the data. If you need to, use 30.5 days in a month.

```python
Activated_data_plot = Activated_data[Activated_data['Activated']< '2018-07-01']
```

```python
Activated_data_plot_sorted = Activated_data_plot.sort_values(by='Activated')
```

```python
Activated_data_plot_sorted['Activated']
```

```python
Activated_data_plot_sorted['month_year'] = pd.to_datetime(Activated_data_plot_sorted['Activated']).dt.to_period('M')
```

```python
Activated_month_year = Activated_data_plot_sorted[['month_year']]
```

```python
Activated_month_year['num'] = 1
activated_sum = Activated_month_year.groupby('month_year').sum()
#activated_sum.reset_index(inplace = True)
```

```python
activated_sum
```

```python
activated_sum.plot(kind= 'bar', stacked= True)
```

```python
activated_sum.reset_index(inplace = True)
activated_sum.reset_index(inplace = True)
activated_sum
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

x = activated_sum[['index']]
y = activated_sum[['num']]
```

```python
model.fit(x, y)

print('intercept:', model.intercept_[0])
print('slope:', model.coef_[0][0])
```
