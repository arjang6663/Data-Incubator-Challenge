```python
df2 = df[['Store Number', 'Item Number', 'Bottles Sold', 'State Bottle Cost', 'State Bottle Retail', 'Date']]
```

```python
df2
```

### Finding out the total profit from each store, for each product:

```python
df2['total_profit'] = (df2['State Bottle Retail'] - df2['State Bottle Cost'])*df2['Bottles Sold']
```

```python
df2 = df2[['Store Number', 'total_profit', 'Date']]
```

### figuring out the total profit from each store:

```python
df3 = df2.groupby('Store Number').sum()
df3
```

### Here we want to find out the total number of orders for each store during this 6 months. Since each row of the dataframe represents order information of a "single" item, we consider all the item orders in each day by each of the stores as a single order. For example, the store with number 2106, had ordered 94 products on 2021-05-06, but we consider it as a single order. Therefore, we need to count the different dates that a store had placed an order, to find out how many times the wholesale had to ship different products to different stores. 

```python
df4 = df2.groupby(['Store Number','Date']).count()
df4.rename(columns={'total_profit': 'order_count'}, inplace=True)
df4
```

### Having the total number of the times each store had placed an order, can help us a lot in different analysis. 
### Relying only on the total profit might be missleading. We need to also know the number of orders for each store. For example, there might a store that make the highest amount of profit for the wholseller, but at the same time it has the highest number of orders. In this case, shipment costs which is not availble in the dataset, may outweight the profit becuase the store has placed too many orders whic made our operation inefficient. 

```python
df4 = df4.groupby('Store Number').count()
df4
```

```python
df3['store_categ'] = df3['total_profit']/ df4['order_count']
df3
```

```python
df3 = df3.sort_values(by='store_categ')
```

```python
df3 = df3.sort_values(by='total_profit')
df3[-5:]
```
